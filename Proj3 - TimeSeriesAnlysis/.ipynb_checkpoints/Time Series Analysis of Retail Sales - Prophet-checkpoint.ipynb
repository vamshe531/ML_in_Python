```python
from fbprophet import Prophet
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("./data/retailsales.csv")
df.head()
```

#### Prophet expects to see the data  which contains a date time field and a y column which we want to forecast 

```python
df.info()
```

df.describe()

#### We need to log transform y variable so that any exponential trends in the data are converted to linear trend and can be modelled using simple linear/additive models

```python
df['y_orig'] = df['y'] #to store original y values
df['y'] = np.log(df['y'])
```

```python
model = Prophet()
model.fit(df)
```

```python
future_data = model.make_future_dataframe(periods=6, freq='m')
```

```python
forecast = model.predict(future_data)
```

```python
model.plot(forecast)
```

```python
model.plot_components(forecast)
```

```python
forecast_bkup=forecast
forecast_bkup['yhat'] = np.exp(forecast_bkup['yhat'])
forecast_bkup['yhat_lower'] = np.exp(forecast_bkup['yhat_lower'])
forecast_bkup['yhat_upper'] = np.exp(forecast_bkup['yhat_upper'])
```

```python
model.plot(forecast_bkup)
```
